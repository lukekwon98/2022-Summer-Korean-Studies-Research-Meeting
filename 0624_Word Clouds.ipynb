## Modify

**Wordcloud Function Parameters**

*   **background_color** = Color of background
*   **max_words** = The maximum number of unique words used, **default value is 200**

*   **stopwords** = stopword list (a, an, are, is, ...) https://www.computerhope.com/jargon/s/stopword.htm

*   **max_font_size** = Maximum font size
*   **random_state** = To ensure that random numbers are generated in the same order, so the results will be the same even if generated several times

*   **width** = width size of the output, **default value is 400**

*   **height** = height size of the output, **default value is 200**
*   **mask** = input an image to be used as a mask

**Modify Colors**

![](https://drive.google.com/uc?export=view&id=12v2UQ6laDBRzCw1_q7OB6uG3scFttEgX)

```python
def one_color_func(word=None, font_size=None, 
                   position=None, orientation=None, 
                   font_path=None, random_state=None):
    h = 160 #Hue (0~360)
    s = 100 #Saturation (0~100)
    l = 50 #Brightness (0~100)
    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
wc = WordCloud(stopwords=STOPWORDS, font_path=font_path,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0], color_func=one_color_func) #add color_func parameter

wc.generate(text)

plt.figure(figsize=(9,9))
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
def similar_color_func(word=None, font_size=None,
                       position=None, orientation=None,
                       font_path=None, random_state=None):
    h = 200 #hue
    s = 100 #saturation
    l = random_state.randint(30, 70) #lightness
    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
wc = WordCloud(stopwords=STOPWORDS, font_path=font_path,
               mask=mask, background_color="white",
               max_words=2000, max_font_size=256,
               random_state=42, width=mask.shape[1],
               height=mask.shape[0], color_func=similar_color_func)
wc.generate(text)

plt.figure(figsize=(9,9))
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```
