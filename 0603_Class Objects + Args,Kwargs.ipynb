# **Creating class "Dog"**

We want to make a variable that can contain all the data of our dogs singlehandedly.

Creating a new data type that is customized for your specific needs.

Initializing basic class : 

```python
class Dog: 
  def __init__(self, name, breed, age, sex): #syntax, must be two underscores on each side
    self.name = name #The dog class will look like this inside
    self.breed = breed 
    self.age = age
    self.sex = sex
    #self.favSnack = snack
    #You can add as many as you like
```

```python
Brie = Dog("Brie", "Golden Retriever", 1, "Female")
print(Brie.name)
print(Brie.breed)
print(Brie.age)
print(Brie.sex)
#print Brie(favSnack)

#All the information about Brie now can be stored in the single variable "Brie", because we  customed a class specifically for that purpose
```

```python
class Dog: 
  def __init__(self, name, breed, age, sex, snack):
    self.name = name
    self.breed = breed 
    self.age = age
    self.sex = sex
    self.favSnack = snack

  def get_info(self): #method = functions you can use on the class in which it is defined. allows you to manipulate properties. Prints out all data in complete sentence form
    if self.sex == "Female":
      print("This dog is a " + str(self.age) + " year old " + self.breed + " and her name is " + self.name+". Her favourite snack is " + self.favSnack + ".")
    elif self.sex == "Male":
      print("This dog is a " + str(self.age) + " year old " + self.breed + " and his name is " + self.name+". His favourite snack is " + self.favSnack + ".")
```

```python
Brie = Dog("Brie", "Golden Retriever", 1, "Female", "Sausages")
Milk = Dog("Milk", "Samoyed", 2, "Female", "Ice Cream")

Brie.get_info()
Milk.get_info()
```

#**Class inheritance**

Inheriting the attributes and methods of one class to another

```python
class Animal: #Create a new simple class. only needs one argument to define name
  def __init__(self, name):
    self.name = name



    

  def print_name(self): #methods = functions using initailized data
    print("The name of this animal is "+self.name)
```

```python
bear = Animal("John")
bear.print_name()
```

```python
class Dog(Animal): #Dog Class inherits Animal Class's traits
  pass
```

```python
Nuggets = Dog("Nuggets")
Nuggets.print_name() #can use methods of class it inherited from. But is useless if inherited this way since it's just copying
```

```python
class Dog(Animal): #Since Animal function only need 'name' attribute which is shared by Dog. We receive all the data we need to initialize Animal class while we initialize Dog class
  def __init__(self, name, breed, age):
    super().__init__(name) #super = superclass(parent class), instead of copy + paste all properties of parent class(Animal)
    self.breed = breed #initalize Dog
    self.age = age

  def get_info(self):
    print("This dog is a " + self.breed + " and its name is " + self.name)
```

As you can see we can get the methods and variables of the parent class by using inheritance

```python
Brie = Dog("Brie", "Golden Retriever", 6)
Brie.print_name() # Dog > Animal Dog inherits all of Animal's methods.
Brie.get_info()
```

**Class Methods & Static Method** **:** **Methods take arguments other than class objects(instance)**

```python
class Dog(Animal):

  animal_name = "Dog" #Automatically becomes a fixed property of class dog, info about the class itself

  def __init__(self, name, breed, age):
    super().__init__(name)
    self.breed = breed
    self.age = age

  def get_info(self): #Functions you use on class objects, Takes pre-defined class object as first argument by default.
    print("This dog is a " + self.breed + " and its name is " + self.name)

  @classmethod #Takes Class (data type) as first argument by default (cls). Function you use on the class itself. cls = class [Dog]
  def animal_type(cls): #when you want to manipulate or use fixed properties of the class itself
    print("This animal is a " + cls.animal_name + " ,from class variable")
    print("This animal is a " + cls.get_animal_type() + " ,from static method") 

  @classmethod
  def unknown_breed(cls, name, age):
    return cls(name, "unknown", age) #takes class Dog as a argument and initializes a class object.

  @staticmethod #when you want info related to the class. don't need a pre-defined object
  def get_animal_type():  #doesn't take arguments at all, to get additional info related to Class
    return "Dog"

  @staticmethod #even though this doesn't take any arguments, it can give info related to class
  def is_adult(age):
    if age > 2:
      print("This dog is an adult")
    else:
      print("This dog is not an adult")
```

Class method could be used to access the class variables. Methods that take Class as argument

```python
Dog.animal_type() #useful when you want to get info about the class itself
```

Class method can also be used to initialize a new class

```python
dog_2 = Dog.unknown_breed("Sam", 3)
# dog2 = Dog("Sam", "unknown", 3)
dog_2.get_info()
```

Use a static method when you don't need to reference the class in any way

```python
Dog.is_adult(3)
```

```python
is_adult(1)
```

Python typing

```python
class Dog(Animal):

  animal_name = "Dog"

  #def __init__(self, name, breed, age):
  def __init__(self, name: str, breed: str, age: int): #hints to datatype of arguments
    super().__init__(name)
    self.breed = breed
    self.age = age

  def get_age(self) -> int: #hints to datatype of returned value
    return self.age
```

You can give type hints to the variables. However, this is not mandatory as Python is a dynamically typed language. This means that the type of a variable is checked in runtime and not compile time. This will throw an error for statically typed languages (Java, C, C++)

```python
dog_3 = Dog("Elizabeth", "Poodle", 9)
dog_3.get_age()
```

```python
dog_3 = Dog("Elizabeth", "Poodle", "9")
dog_3.age

#more of a guideline than a rule
```

Using *args.

```python
def print_class(teacher, student1, student2, student3): #A function that prints all of the members in a classroom
  print("The teacher is: " + teacher)
  print("And the students are: " + student1+", " + student2+", " + student3)
```

*args : 

"*" = all (*.txt), "arg" = arguments. 

-> Take all arguments

Useful when you don't know how many students(arguments) there are. Passes unspeicified number of variables as a tuple


print() function uses *args

https://www.programiz.com/python-programming/methods/built-in/print


```python
def print_class(teacher, *args): 
  print("The teacher is: " + teacher)
  print("And the students are: " + ', '.join(args))

  print("\n")
  print(args)

print_class("John", "Tom", "Michael", "Emily", "Anne", "Brian", "Leo", "Benjamin", "Lucy", "James")
```

Using **kwargs you can pass in variables with an assigned key. 

Useful when there are too many arguments for a function but you only need to change a few.

```python
def print_leader(**kwargs):
  for key, value in kwargs.items():
    if key == "Leader":
      print("The " + key + " is " + value)
  
  print(kwargs) #prints in dict form, key : value
  print(kwargs['Leader']) #you can access each individual value by key
```

```python
print_leader(
    Leader = "Tom",
    Member_1 = "Chris",
    Member_2 = "Christina"
)
```

```python
print_leader(
    Member_1 = "Chris",
    Leader = "Tom",
    Member_2 = "Christina"
) #Leader is still Tom even though values are given in different order
```

```python
team = {'Leader': 'Tom', 'Member_1': 'Chris', 'Member_2': 'Christina'}
print_leader(**team) #you can also give a whole dict type as argument
#example = KSDS visualization
```
